# 03 — Gráficos de espectros

`ManimGraph` envuelve la creación de ejes, graficado de datos desde
archivos y leyendas en una sola clase.

## Formato esperado

Archivo de 2 columnas (x, y) sin encabezado, separado por espacios o
tabulaciones. Si tiene encabezado, usa `pandas` (incluido como fallback).

## Requisitos

- `numpy`
- `scipy` (para `savgol_filter` cuando `smooth=True`)

```python
import sys
sys.path.insert(0, '..')

import numpy as np

from manim import *
from manim_slides import Slide
from toolkit import ManimGraph
```

```python
# Generar datos sintéticos para probar sin un archivo real
x = np.linspace(300, 800, 200)
y = np.exp(-((x - 500) ** 2) / (2 * 30 ** 2)) + 0.1 * np.random.random(200)
np.savetxt("/tmp/espectro_demo.txt", np.column_stack([x, y]))
print("Datos guardados en /tmp/espectro_demo.txt")
```

```python
%%manim_slides -qm DemoEspectro

class DemoEspectro(Slide):
    def construct(self):
        helper = ManimGraph(self)
        axes = helper.setup_axes(
            x_label=r"\lambda \,(\mathrm{nm})",
            y_label="Absorbancia",
            x_range=[300, 800, 50],
            y_range=[0, 1.2, 0.2],
        )
        self.play(Create(axes))
        self.next_slide()

        curva_cruda = helper.plot_spectrum("/tmp/espectro_demo.txt", color=BLUE, smooth=False)
        self.play(Create(curva_cruda))
        self.next_slide()

        curva_suave = helper.plot_spectrum("/tmp/espectro_demo.txt", color=GREEN, smooth=True)
        self.play(Create(curva_suave))
        self.next_slide()

        leyenda = helper.create_legend([
            {"text": "Crudo", "color": BLUE},
            {"text": "Suavizado (Savitzky-Golay)", "color": GREEN},
        ], position=UR)
        self.play(Write(leyenda))
        self.next_slide()
        self.wait()
```
