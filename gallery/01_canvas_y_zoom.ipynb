# 01 — Canvas persistente y ZoomedScene

`SlidesControl` combina `Slide` y `ZoomedScene` para ofrecer:

- Canvas con título y número de diapositiva persistentes.
- `update_canvas()` para incrementar el número visible.
- `clear_slide_content()` para limpiar el resto.
- Herencia de `ZoomedScene` para hacer zoom a regiones.

## Ejecutar

Renderiza con `manim-slides render`:

```bash
manim-slides render gallery/01_canvas_y_zoom.ipynb DemoCanvasZoom -ql
manim-slides DemoCanvasZoom
```

Para la versión con magic de Jupyter, usa `%%manim_slides`.

````python
import sys
sys.path.insert(0, '..')

from manim import *
from manim_slides import Slide
from toolkit import SlidesControl, NORMAL_SIZE, TITLE_SIZE
````

````python
%%manim_slides -qm DemoCanvasZoom

class DemoCanvasZoom(SlidesControl):
    def construct(self):
        # Canvas persistente: título arriba-izquierda, número abajo-izquierda
        titulo = Title("Canvas y Zoom", font_size=TITLE_SIZE).to_corner(UL)
        num = Text("1").to_corner(DL)
        self.add_to_canvas(title=titulo, slide_number=num)

        self.play(Write(titulo), Write(num))
        self.next_slide()

        # Contenido principal con varios mobjects
        circulo = Circle(radius=1.5, color=BLUE).shift(LEFT * 3)
        cuadrado = Square(side_length=2, color=GREEN).shift(RIGHT * 3)
        self.play(Create(circulo), Create(cuadrado))
        self.next_slide()

        # ZoomedScene heredado: zoom al cuadrado
        zoomed_camera = self.zoomed_camera
        zoomed_display = self.zoomed_display
        zoomed_display.move_to(RIGHT * 3)

        self.activate_zooming()
        self.play(self.get_zoom_in_animation(), run_time=2)
        self.wait()
        self.next_slide()

        # Salir del zoom
        self.play(self.get_zoom_out_animation(), run_time=2)
        self.next_slide()

        # Limpiar contenido (título y número permanecen)
        self.clear_slide_content()
        self.wait()
````
