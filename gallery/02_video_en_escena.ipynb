# 02 — Video dentro de una escena

`VideoMobject` envuelve un video como `ImageMobject` y lo actualiza
frame a frame mediante un updater. Soporta control de velocidad y loop.

## Requisitos

- `opencv-python` (`pip install opencv-python`)
- `Pillow`
- Un archivo de video en `figures/` (puedes usar el de la defensa).

## Notas

- Si el video es muy pesado, el rendering será lento. Usa `-ql` para iterar.
- En `defensa_final.py` se desactiva `skip_reversing` para slides con video.

```python
import sys
sys.path.insert(0, '..')

from manim import *
from manim_slides import Slide, ManimSlidesMagic 
from toolkit import VideoMobject
```

```python
import os

# Busca el primer .mp4 disponible en figures/
FIGURES_DIR = os.path.join('..', 'figures')
video_path = None
if os.path.isdir(FIGURES_DIR):
    for fname in sorted(os.listdir(FIGURES_DIR)):
        if fname.lower().endswith(('.mp4', '.mov', '.avi')):
            video_path = os.path.join(FIGURES_DIR, fname)
            break

print(f'Video encontrado: {video_path if video_path else "NINGUNO (usar placeholder)"}')
```

```python
class DemoVideo(Slide):
    def construct(self):
        titulo = Title("Video en escena").to_corner(UL)
        self.play(Write(titulo))
        self.next_slide()

        if video_path is not None:
            # Reproducir el video encontrado
            video = VideoMobject(video_path, speed=1.0, loop=True)
            video.height = 5
            self.play(FadeIn(video))
            self.next_slide(loop=True)  # Loop hasta que el usuario avance
            self.wait(3)
            self.next_slide()
            self.play(FadeOut(video))
        else:
            # Placeholder si no hay video
            placeholder = Text(
                "Coloca un .mp4 en figures/\\npara ver este demo"
            )
            self.play(Write(placeholder))
            self.next_slide()
            self.play(FadeOut(placeholder))

        self.wait()
```
