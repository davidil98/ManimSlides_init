```python
import manim as mn
from manim import *

config.media_width = "75%"
config.verbosity = "WARNING"

print(mn.__version__)
```

```python
%%manim -qm Escena1_Introduccion

class Escena1_Introduccion(Scene):
    def construct(self):
        # Creamos el objeto
        circuito = Circle().scale(2)
        self.play(Create(circuito))
        self.wait()
        # Al final de esta escena, el circuito se queda fijo en pantalla

```

```python
%%manim -qm [Escena1_Introduccion, Escena2_Funcionamiento]

class Escena2_Funcionamiento(Scene):
    def construct(self):
        # Replicamos exactamente el mismo objeto en su posición final
        circuito = Circle().scale(2)
        
        # El truco: Usamos self.add() en lugar de self.play()
        # Esto lo coloca en la pantalla instantáneamente desde el frame 0
        self.add(circuito)
        
        # Ahora la audiencia siente que es el mismo objeto, 
        # pero tú estás trabajando en un archivo de código limpio y nuevo
        explicacion = Text("Flujo de iones en el canal...")
        self.play(Write(explicacion))
        self.wait()
```
